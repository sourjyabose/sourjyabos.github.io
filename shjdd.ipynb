```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

data=pd.read_csv("train2.csv")

data.head(100)


data=np.array(data)

m,n=data.shape
np.random.shuffle(data)
dev=data[1:1000].T
ydev=dev[0]
xdev=dev[1:n]/255
train=data[1000:m].T
ytrain=train[0]
xtrain=train[1:n]/255




def init_param():
    w1=np.random.rand(10,784)-0.5
    b1=np.random.rand(10,1)-0.5
    w2=np.random.rand(10,10)-0.5
    b2=np.random.rand(10,1)-0.5
    return w1,w2,b1,b2

def Relu(Z):
    return np.maximum(0,Z)

def softMax(p1):
    return np.exp(p1)/np.sum(np.exp(p1),axis=0)

def forward_prop(w1,b1,w2,b2,X):
    Z1=w1.dot(X)+b1
    
    A1=Relu(Z1)
    Z2=w2.dot(A1)+b2
    A2=softMax(Z2)
    return Z1,A1,Z2,A2

def one_hot(Y):
    one_hot=np.zeros((Y.size,Y.max()+1));
    one_hot[np.arange(Y.size),Y]=1
    return one_hot.T

def dRelu(x):
    return x>0
        
def back_prop(Z1,A1,Z2,A2,Y,W2,X):
    m=Y.size
    onehot=one_hot(Y)
    dZ2=A2-onehot
    dW2=1/m*dZ2.dot(A1.T)
    dB2=1/m*np.sum(dZ2,axis=1,keepdims=True)
    dZ1=W2.T.dot(dZ2)*dRelu(Z1)
    dW1=1/m*dZ1.dot(X.T)
    dB1=1/m*np.sum(dZ1,axis=1,keepdims=True)
    return dZ2,dW2,dB2,dZ1,dW1,dB1

def update_param(w1,w2,b1,b2,dw1,dw2,db1,db2,alpha):
    w1=w1-alpha*dw1
    b1=b1-alpha*db1
  
    w2=w2-alpha*dw2
    b2=b2-alpha*db2
    return w1,b1,w2,b2

def get_pred(A2):
    return np.argmax(A2,0)

def get_accuracy(prediction,y):
    print(prediction,y)
    return np.sum(prediction==y)/y.size


def gradient_desc(x,y,itera,alpha):
    w1,w2,b1,b2=init_param()
    print(x.shape)
    for i in range(itera):
        
       
        Z1,A1,Z2,A2=forward_prop(w1,b1,w2,b2,x)
        dZ2,dW2,dB2,dZ1,dW1,dB1=back_prop(Z1,A1,Z2,A2,y,w2,x)
        w1,b1,w2,b2=update_param(w1,w2,b1,b2,dW1,dW2,dB1,dB2,alpha)
        if(i%10==0):
            print("\n-------\n")
            print("Round: ",i)
            print("Accuracy",get_accuracy(get_pred(A2),y))
    return w1,w2,b1,b2

w1,w2,b1,b2=gradient_desc(xtrain,ytrain,500,0.1)
```
